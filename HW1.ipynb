```julia
#p5, b)
R = Sym("R");V = Sym("V"); Rd= Sym("Rd");Rb=Sym("Rb");
function dV0(R,V,Rd,Rb)
    V*R*(1/(Rb+R)-1/(Rd+R));
end
f = dV0(R,V,Rd,Rb)
dV0_dR=diff(f,R);
R=solve(dV0_dR,R)
```
