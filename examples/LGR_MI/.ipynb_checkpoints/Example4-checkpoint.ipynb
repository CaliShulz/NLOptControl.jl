This example is actually identical to Example3 except we push some of the functionality into the Code.

The idea, is that it can be easily used when optimizing. So, you pass it the states and controls etc. and it calculates the rest.


```julia
using NLOptControl
using Polynomials
using Plots
using FastGaussQuadrature
pyplot()
default(guidefont = font(17), tickfont = font(15), legendfont = font(12), titlefont = font(20))
```

```julia
# define problem
t0 = Float64(0); tf = Float64(10);  # TODO change and y to x
t = Array(linspace(t0,tf,100));
α₁ =  -0.3; α₂ = 3; α₃ = -8; α₄ =  7;
Nc = Int64(3); # number of collocation points in each interval
Ni = Int64(2);  # number of intervals
####################################
# perform analytical calcualtions
####################################
γ = Poly([α₁,α₂,α₁]); #TODO check on that imported binding warning
y = polyval(γ,t);

# evaluate the integral
∫γ = polyint(γ);
Y = polyval(∫γ,t[end]) - polyval(∫γ,t[1]);
C = Y - polyval(∫γ,t[end]); # constant of integration
∫y = polyval(∫γ,t) + C;

# evaluate the derivative
dγ = polyder(γ);
dy = polyval(dγ,t);

# the final time (tf) is divided by the number of intervals and these are the points where we have dvs.
# the collocation points within this region are created using LGR nodes. But, if there is not state data here, 
# how do we approximate? 

# TODO may be able to calculate this ahead of time
τ, ω = gaussradau(Nc); # number of collocation points per interval

# break the problem up into multiple intervals
di, tm, t_data, ωₛ = create_intervals(t0,tf,Ni,Nc,τ,ω); # TODO

# append final node τ=+1
temp = zeros(Float64,1,Ni);  # we are leaving this out of create_intervals() because not all method require this appendage
for int in 1:Ni
temp[int] = di*int;
end
t_data = [t_data;temp];

# this will be the design variables, i.e. states and controls
# TODO create functionality to break variables up into intervals
x = zeros(Float64,Nc+1,Ni); 
for int in 1:Ni
  x[:,int] = polyval(γ,t_data[:,int]);
end
    
###############################################
# use new functionality for taking derivatives
###############################################
xd=lgr_diff(Nc,Ni,x,t_data)

###############
# post processing
#################
dp=plot(0,leg=:false)
plot!(t,dy,label="derivative",w=6)
for int in 1:Ni
    scatter!(t_data[1:end-1,int],xd[:,int],markersize =10,markershape = :rect,label=string("approximate derivative ",int))
end
xlims!(t0,tf*1.1)

plot(dp)
```
